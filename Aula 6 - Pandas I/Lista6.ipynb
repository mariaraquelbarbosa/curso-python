```python
verificar_hashtag = pd.DataFrame(tiktok_ordem['video_desc'].str.contains(pat = '#fyp') == 1)
verificar_hashtag_ID = (pd.DataFrame(verificar_hashtag).join(tiktok['video_id'],how="outer"))
proporcao_bool = verificar_hashtag_ID.groupby("video_desc").count()[["video_id"]]
proporcao_bool
```

### Análises e conclusões obtidas
Primeiramente, é importante pontuar que os posts foram alocados em ordem decrescente de acordo com o número total de interações, calculado a partir da soma de curtidas, comentários, compartilhamentos e reproduções. Depois, foi possível notar que todos os 6 primeiros posts com maior número de interações apresentaram a hashtag "#fyp" em sua descrição. Além disso, analisando a base como um todo, a qual já engloba vídeos que apresentam grande popularidade no TikTok, a incidência de posts que apresentam a mesma hashtag é de 15/18 (83,3%), podendo assim ser observada uma influência dessa hashtag no aumento do número de interações dos vídeos nessa rede social.

Para mais, analisando estatísticas descritivas, nota-se que a média de tempo dos 5 vídeos com maior interação na base analisada (29,4 s) é maior que a média de tempo da base como um todo (16,8 s). Além disso, a média de tempo dos 5 vídeos com menor interação da base analisada (11,2 s) é menor que a média de tempo da base como um todo (16,8 s). A partir disso, é possível concluir também que, em média, as pessoas tendem a interagir mais com vídeos que apresentam maior tempo de duração.
