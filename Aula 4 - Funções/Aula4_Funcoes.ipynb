# Escopo de funções e funçõe alinhadas

## Diferença entre variável global e local

```python
a = 'variável global'
#O a ele vale como 'variável global' em todo o código

def func():
    a = "variável local"
    print(a)

func()
#Dentro dessa função, o a vale como 'variável local'
```

```python
print(a)
```

```python
def funcao():
    global a
    a = "Alterando a Variável global dentro da função"
    b = "Outra variável local"
    print(a)
    
funcao()
```

```python
print(a)
#Não é mais "variável global", mas sim "Alterando a Variável global dentro da função"
```

```python
print(b)
#Não é uma variável global
```

## Funçõoes alinhadas

Funções dentro de funções

```python
#aqui, não importa a ordem que definimos a função
def produto(x,y):
    return x*y

def quadrado_soma(a,b):
    return a**2 + 2*produto(a,b) + b**2
```

```python
quadrado_soma(2,3)
```

```python
def main():
    name = input('Qual o seu nome? ')
    hello(name)

def hello(name):
    print('Hello', name)
main()
```

# Múltiplos retornos e argumentos opcionais

```python
def operacoes(a,b):
    soma = a+b
    diferenca = a-b
    prod = a*b
    div = a/b
    return soma, diferenca, prod, div
```

```python
operacoes(2,3)
```

```python
soma, diferenca, prod, div = operacoes(2,8)
print(f'A soma é {soma}, a diferença é {diferenca}, o produto é {prod} e o quociente é {div}')
```
